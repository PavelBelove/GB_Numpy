–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ª–∏–Ω–µ–π–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É, –∏—Å–ø–æ–ª—å–∑—É—è –º–∞—Å—Å–∏–≤ t –¥–ª—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏,–∞ –º–∞—Å—Å–∏–≤ f - –¥–ª—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏. –õ–∏–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –∑–µ–ª–µ–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞.
–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã - '–ì—Ä–∞—Ñ–∏–∫ f(t)'. –¢–∞–∫–∂–µ –¥–æ–±–∞–≤—å—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏—è –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –æ—Å–∏ - '–ó–Ω–∞—á–µ–Ω–∏—è t' –∏ –¥–ª—è –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π - '–ó–Ω–∞—á–µ–Ω–∏—è f'.
–û–≥—Ä–∞–Ω–∏—á—å—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –ø–æ –æ—Å–∏ x –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ 0.5 –∏ 9.5, –∞ –ø–æ –æ—Å–∏ y - –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ -2.5 –∏ 2.5.


```python
t = np.linspace(0, 10, 51)
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#505050",
    "family": "serif",
}
axis_font = {
    "fontsize": 12,
    "fontweight": "bold",
    "color": "darkgray",
    "family": "serif",
}
```

```python
f = np.cos(t)
```

```python
f_from_t = pd.DataFrame(f, t)
```

```python
f_from_t.plot(color = 'green')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.title('–ì—Ä–∞—Ñ–∏–∫ f(t)', fontdict = title_font)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è t', fontdict = axis_font)
plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏—è f', fontdict = axis_font);
```

```python
DATASET_CREDIT_CARD = '../data/creditcard.csv'
```

```python
cdc = pd.read_csv(DATASET_CREDIT_CARD)
```

```python
counts_class = cdc['Class'].value_counts()
```

```python
counts_class.plot(kind="barh", style='fivethirtyeight', logx=True);
```

```python
plt.hist(cdc[cdc['Class'] == 0]['V1'], bins=20, ec='white', color='gray', alpha=0.5, density=True)
plt.hist(cdc[cdc['Class'] == 1]['V1'], bins=20, ec='white', color='red', alpha=0.5, density=True)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö –∏ –º–æ—à–µ–Ω–Ω–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É X1')
plt.xlabel('Class')
plt.legend(labels=["Normal", "Scam"]);
```

9) –°–æ–∑–¥–∞—Ç—å –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ –∞, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 12. –ü–æ–º–µ–Ω—è—Ç—å —Ñ–æ—Ä–º—É —ç—Ç–æ–≥–æ –º–∞—Å—Å–∏–≤–∞, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏–ª–∞—Å—å –º–∞—Ç—Ä–∏—Ü–∞ A (–¥–≤—É–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ Numpy), —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ 4 —Å—Ç—Ä–æ–∫ –∏ 3 —Å—Ç–æ–ª–±—Ü–æ–≤. –ü–æ–ª—É—á–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—É At –ø—É—Ç–µ–º —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã A. –ü–æ–ª—É—á–∏—Ç—å –º–∞—Ç—Ä–∏—Ü—É B, —É–º–Ω–æ–∂–∏–≤ –º–∞—Ç—Ä–∏—Ü—É A –Ω–∞ –º–∞—Ç—Ä–∏—Ü—É At —Å –ø–æ–º–æ—â—å—é –º–∞—Ç—Ä–∏—á–Ω–æ–≥–æ —É–º–Ω–æ–∂–µ–Ω–∏—è. –ö–∞–∫–æ–π —Ä–∞–∑–º–µ—Ä –∏–º–µ–µ—Ç –º–∞—Ç—Ä–∏—Ü–∞ B? –ü–æ–ª—É—á–∏—Ç—Å—è –ª–∏ –≤—ã—á–∏—Å–ª–∏—Ç—å –æ–±—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã B –∏ –ø–æ—á–µ–º—É?

```python
a = np.arange(0, 12)
A = a.reshape(4, 3)
At = A.transpose()
```

```python
B = A.dot(At)
# np.linalg.inv(B) # –û—à–∏–±–∫–∞ - —Å–∏–Ω–≥—É–ª—è—Ä–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞
B
```

11) –°–æ–∑–¥–∞–π—Ç–µ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤ c, —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–π –∏–∑ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ 16-—Ç–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –æ—Ç 0 –¥–æ 16 –Ω–µ–≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ.

```python
def generate():
    for x in range(16):
        yield np.random.randint(16)
```

```python
a = np.fromiter(generate(), dtype='int')
a
```

12) –ü–æ–º–µ–Ω—è–π—Ç–µ –µ–≥–æ —Ñ–æ—Ä–º—É —Ç–∞–∫, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏–ª–∞—Å—å –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ C. –ü–æ–ª—É—á–∏—Ç–µ –º–∞—Ç—Ä–∏—Ü—É D, –ø–æ—ç–ª–µ–º–µ–Ω—Ç–Ω–æ –ø—Ä–∏–±–∞–≤–∏–≤ –º–∞—Ç—Ä–∏—Ü—É B –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –≤–æ–ø—Ä–æ—Å–∞ –∫ –º–∞—Ç—Ä–∏—Ü–µ C, —É–º–Ω–æ–∂–µ–Ω–Ω–æ–π –Ω–∞ 10. –í—ã—á–∏—Å–ª–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å, —Ä–∞–Ω–≥ –∏ –æ–±—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É D_inv –¥–ª—è D.

```python
C = a.reshape(4, -1)
```

```python
D = C * 10 + B
D
```

```python
np.linalg.det(D)
```

```python
np.linalg.matrix_rank(D)
```

```python
D_inv = np.linalg.inv(D)
```

13) –ü—Ä–∏—Ä–∞–≤–Ω—è–π—Ç–µ –∫ –Ω—É–ª—é –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –º–∞—Ç—Ä–∏—Ü–µ D_inv, –∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ - –∫ –µ–¥–∏–Ω–∏—Ü–µ. –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤ –º–∞—Ç—Ä–∏—Ü–µ D_inv –æ—Å—Ç–∞–ª–∏—Å—å —Ç–æ–ª—å–∫–æ –Ω—É–ª–∏ –∏ –µ–¥–∏–Ω–∏—Ü—ã. –° –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ numpy.where, –∏—Å–ø–æ–ª—å–∑—É—è –º–∞—Ç—Ä–∏—Ü—É D_inv –≤ –∫–∞—á–µ—Å—Ç–≤–µ –º–∞—Å–∫–∏, –∞ –º–∞—Ç—Ä–∏—Ü—ã B –∏ C - –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ –¥–∞–Ω–Ω—ã—Ö, –ø–æ–ª—É—á–∏—Ç–µ –º–∞—Ç—Ä–∏—Ü—É E —Ä–∞–∑–º–µ—Ä–æ–º 4x4.  –≠–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã E, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ã D_inv —Ä–∞–≤–µ–Ω 1, –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–≤–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–º—É —ç–ª–µ–º–µ–Ω—Ç—É –º–∞—Ç—Ä–∏—Ü—ã B, –∞ —ç–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç—Ä–∏—Ü—ã E, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π —ç–ª–µ–º–µ–Ω—Ç –º–∞—Ç—Ä–∏—Ü—ã D_inv —Ä–∞–≤–µ–Ω 0, –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–≤–Ω—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–º—É —ç–ª–µ–º–µ–Ω—Ç—É –º–∞—Ç—Ä–∏—Ü—ã C.


```python
D_inv[D_inv > 0] = 1
D_inv[D_inv < 0] = 0
D_inv
```

```python

E = np.where(D_inv, B, C)
E
```
