```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
plt.hist(cdc[cdc['Class'] == 0]['V1'], bins=20, ec='white', color='gray', alpha=0.5, density=True)
plt.hist(cdc[cdc['Class'] == 1]['V1'], bins=20, ec='white', color='red', alpha=0.5, density=True)
plt.title('Распределение нормальных и мошеннических операций по признаку X1')
plt.xlabel('Class')
plt.legend(labels=["Normal", "Scam"]);
```
